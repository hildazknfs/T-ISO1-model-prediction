### Visualisasi Model

```python
import joblib
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree, export_text
```

```python
rf_model_filename = r"D:\HildasCollegeStuffs\PROPOSAL\Model\random_forest_model.pkl"
dt_model_filename = r"D:\HildasCollegeStuffs\PROPOSAL\Model\decision_tree_model.pkl"

trained_model_rf = joblib.load(rf_model_filename)
trained_model_dt = joblib.load(dt_model_filename)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Ambil 5 pohon pertama dari Random Forest
n_trees_to_show = 5
trees = trained_model_rf.estimators_[:n_trees_to_show]

plt.figure(figsize=(25, 5))  # Lebar besar untuk 5 pohon sejajar, tinggi kecil

for i, tree in enumerate(trees):
    plt.subplot(1, n_trees_to_show, i + 1)  # 1 baris, 5 kolom
    plot_tree(
        tree,
        filled=True,
        feature_names=['t (s)', 'T-iSO2'],
        rounded=True,
        fontsize=6  # font lebih kecil supaya muat
    )
    plt.title(f"Pohon ke-{i+1}")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plot_tree(trained_model_dt, filled=True, feature_names=['t (s)', 'T-iSO2'], rounded=True, fontsize=8)
plt.title("Struktur Decision Tree", fontsize=8)
plt.show()

rules_dt = export_text(trained_model_dt, feature_names=['t (s)', 'T-iSO2'])
print("Aturan Decision Tree:")
print(rules_dt)
```
